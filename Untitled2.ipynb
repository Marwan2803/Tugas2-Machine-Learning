```python
import pandas as pd

# Buat data dalam bentuk dictionary
data = {
    "Pendapatan_rata2_1000$": [5, 10, 20, 8, 4, 6, 12, 15],
    "Penjualan_Pizza_1000_buah": [27, 46, 73, 40, 30, 28, 46, 59]
}

# Konversi ke DataFrame
df = pd.DataFrame(data)

# Simpan ke file CSV
df.to_csv("penjualan_pizza.csv", index=False)

print("File CSV berhasil dibuat dan disimpan!")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Baca file CSV yang sudah dibuat
df = pd.read_csv("penjualan_pizza.csv")

# Scatter plot
plt.scatter(df["Pendapatan_rata2_1000$"], df["Penjualan_Pizza_1000_buah"], color='blue')
plt.xlabel("Pendapatan Rata-rata (1000$)")
plt.ylabel("Penjualan Pizza (1000 buah)")
plt.title("Scatter Plot Pendapatan vs Penjualan Pizza")
plt.show()

# Regresi linear
X = df["Pendapatan_rata2_1000$"].values.reshape(-1, 1)
y = df["Penjualan_Pizza_1000_buah"].values
model = LinearRegression()
model.fit(X, y)

# Koefisien regresi
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Prediksi nilai y
y_pred = model.predict(X)

# Menghitung SSR, SSE, SST
SSR = np.sum((y_pred - y.mean())**2)
SSE = np.sum((y - y_pred)**2)
SST = SSR + SSE
R2 = r2_score(y, y_pred)

print(f"SSR: {SSR}")
print(f"SSE: {SSE}")
print(f"SST: {SST}")
print(f"Koefisien Determinasi (RÂ²): {R2}")
```
